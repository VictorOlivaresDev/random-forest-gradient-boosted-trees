```python
# Importación de librerías necesarias
import pandas as pd
from sklearn. preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Carga del DataFrame
drugs = pd.read_csv('drugs.csv')
drugs
```

```python
# Creación de variable que contiene columnas a utilizar
feature_cols = ['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']
```

```python
# Separaciíon de datos predictores y variable a predecir
x = drugs[feature_cols].values
x
```

```python
y = drugs.Drug
y
```

```python
# Codificación de variables categóricas
code_sex = LabelEncoder()
code_sex.fit(['M', 'F'])
x[:,1] = code_sex.transform(x[:,1])
x
```

```python
code_bp = LabelEncoder()
code_bp.fit(['LOW', 'NORMAL', 'HIGH'])
x[:,2] = code_bp.transform(x[:,2])
x
```

```python
code_cholesterol = LabelEncoder()
code_cholesterol.fit(['NORMAL', 'HIGH'])
x[:,3] = code_cholesterol.transform(x[:,3])
x
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Creación de grupos de prueba y entrenamiento
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
# Creación del modelo
rf = RandomForestClassifier(n_estimators=100, random_state=1)
rf.fit(x_train, y_train)
```

```python
# Creación de variable de predicción y Reporte de Clasificación
y_pred = rf.predict(x_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
# Creación de función para graficar la importancia de las variables de acuerdo al modelo dado
def plot_importances(df, model):
    n_cols = feature_cols
    plt.barh(n_cols, model.feature_importances_, align='center')
    plt.xlabel('Importancia de las variables')
    plt.ylabel('Características')
    plt.show()
    
```

```python
# Gráfica de Random Forest
plot_importances(drugs, rf)
```

# Gradient Boost Trees

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# Creación del Modelo Gradient Boost Tree
gbt = GradientBoostingClassifier(n_estimators=100, random_state=1)
gbt.fit(x_train, y_train)
```

```python
# Predicción del modelo y Reporte de Clasificación
y_pred = gbt.predict(x_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
# Generación del gráfico de Gradient Boost Tree
plot_importances(drugs, gbt)
```

# AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier(n_estimators=100, random_state=1)
abc.fit(x_train, y_train)
```

```python
# Predicción y Reporte de Clasificación
y_pred = abc.predict(x_test)
print(metrics.classification_report(y_test, y_pred))
```
